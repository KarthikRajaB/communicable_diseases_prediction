# Correlation Check between the target variable and other attributes

Correlation matrix - Simply said, a correlation matrix is a table that shows the correlation coefficients for various variables. 
The correlation between all potential pairs of values in a table is shown in the matrix. 
It is an effective tool for compiling a sizable dataset and for locating and displaying data patterns.

```python
plt.figure(figsize = (15,9))
corelation = raw_data.corr()
corelation
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
index = 0
monthYears = []
years= [i for i in raw_data.year]
for i in raw_data.month:
    monthYears.append(str(years[index])+'-'+str(i))
    index += 1
plt.figure(figsize=(14,6))
sns.lineplot(x=monthYears,y=[i for i in raw_data['cases']],color='black')
plt.title(label='Dengue fever cases from 2011-2021')
plt.xlabel(xlabel='Month')
plt.ylabel(ylabel='Cases')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (14,6))
raw_data.groupby('year')['cases'].sum().plot.bar()
```

# The above bar graph indicates the yearwise count of dengue cases from 2011 to 2021

```python
plt.figure(figsize = (14,10))

raw_data.corr().abs().cases.drop('cases').sort_values().plot.barh(color='darkblue')
plt.title('Variables Correlation with cases variables')
```

# The above graph indicates the correlation between the case count and other parameters

```python
plt.figure(figsize=(17,8))
raw_data.plot(x="cases", y="prec", kind="scatter")
```

```python
def compCase (year):
    year = int(year)
    year_20xx = raw_data.loc[raw_data['year'] == year]
    
    fig,a =  plt.subplots(2, figsize=(13,5), gridspec_kw={'height_ratios': [1, 2]})
    a[0].plot(year_20xx["prec"])
    a[0].set_title('RAINFALL')
    a[1].plot(year_20xx["cases"])
    a[1].set_title('CASES')
    fig.suptitle('Year '+ str(year), fontsize=20)
    plt.show()
```
