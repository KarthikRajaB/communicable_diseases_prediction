```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
%matplotlib inline
```

```python
pwd = os.getcwd()
filepath = os.path.join(pwd,"DATA/cg.csv")
filepath
```

```python
raw_data = pd.read_csv(filepath)
raw_data
```

```python
plt.figure(figsize = (15,9))
corelation = raw_data.corr()
corelation
sns.heatmap(corelation, xticklabels=corelation.columns, yticklabels=corelation.columns, annot=True)
```

```python
index = 0
monthYears = []
years= [i for i in raw_data.year]
for i in raw_data.month:
    monthYears.append(str(years[index])+'-'+str(i))
    index += 1
plt.figure(figsize=(14,6))
sns.lineplot(x=monthYears,y=[i for i in raw_data['cases']],color='black')
plt.title(label='Dengue fever cases from 2011-2021')
plt.xlabel(xlabel='Month')
plt.ylabel(ylabel='Cases')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (14,6))
raw_data.groupby('year')['cases'].sum().plot.bar()
```

```python
plt.figure(figsize = (14,10))

raw_data.corr().abs().cases.drop('cases').sort_values().plot.barh(color='darkblue')
plt.title('Variables Correlation with cases variables')
```

```python
plt.figure(figsize=(17,8))
(raw_data
     .cases
     .plot
     .line(lw=0.8))
plt.title('Dengue cases in Kerala ', fontsize = 25)
plt.xlabel('Time',fontsize = 20)
```

```python
plt.figure(figsize=(17,8))
(raw_data
     .temp_max
     .plot
     .line(lw=0.8))

plt.title('Temprature over Time')
plt.xlabel('Time')
```

```python
def compCase (year):
    year = int(year)
    year_20xx = raw_data.loc[raw_data['year'] == year]
    
    fig,a =  plt.subplots(2, figsize=(13,5), gridspec_kw={'height_ratios': [1, 2]})
    a[0].plot(year_20xx["temp_max"])
    a[0].set_title('Temperature')
    a[1].plot(year_20xx["cases"])
    a[1].set_title('CASES')
    fig.suptitle('Year '+ str(year), fontsize=20)
    plt.show()
```
